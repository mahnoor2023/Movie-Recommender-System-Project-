```python
import ast

def convert(text):
    if isinstance(text, list):   # already converted
        return text
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```
